### Density and Contour Plots ###

Sometimes it is useful to display three-dimensional data in two dimensions using
contours or color-coded regions. There are three Matplotlib functions that can be
helpful for this task: plt.contour for contour plots, plt.contourf for filled contour
plots, and plt.imshow for showing images. This section looks at several examples of
using these. Well start by setting up the notebook for plotting and importing the
functions we will use:

**Visualizing a Three-Dimensional Function**

```python
# z = f( x, y)
def f(x,y):
    return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)
```

```python
x = np.linspace(0,10,50)
y = np.linspace(0,10,40)
```

```python
X, Y = np.meshgrid(x,y)
```

```python
Z = f(X, Y)
```

```python
plt.contour(X, Y, Z)
```

```python
plt.contour(X, Y, Z, cmap = 'plasma')
```

```python
plt.contourf(X, Y, Z, cmap = 'plasma')
plt.colorbar()
```

```python
contours = plt.contour(X, Y, Z, colors='black')
plt.clabel(contours, inline =True)
plt.imshow(Z, cmap='plasma',alpha=0.3,extent=[0,5,0,5])
plt.colorbar()
```

**Plot Customization by Hand**


```python
x = np.random.randn(1000)
plt.hist(x)
```

```python
ax = plt.axes()
ax.set_axisbelow(True)

# draw solid white grid lines
plt.grid(color='w', linestyle='solid')
    
# hide axis spines
for spine in ax.spines.values():
    spine.set_visible(False)
    
# hide top and right ticks
ax.xaxis.tick_bottom()
ax.yaxis.tick_left()

# lighten ticks and labels
ax.tick_params(colors='gray', direction='out')
for tick in ax.get_xticklabels():
    tick.set_color('gray')
for tick in ax.get_yticklabels():
    tick.set_color('gray')

# control face and edge color of histogram
ax.hist(x, edgecolor='#E6E6E6', color='#EE6666');
```

### Lets create a function that will make two basic types of plot:

```python
def hist_and_line():
    np.random.seed(0)
    fig, ax = plt.subplots(1, 2,figsize=(22,9))
    ax[0].hist(np.random.randn(1000))
    for i in range(4):
        ax[1].plot(np.random.randn(10))
    ax[1].legend(['a','b','c','d'],loc = 'lower left', fontsize = 15)
    
```

```python
hist_and_line()
```

```python
# we should set default style
plt.rcParams.update(Ipython_default)
```

```python
hist_and_line()
```

```python
with plt.style.context('fivethirtyeight'):
    hist_and_line()
```

```python
with plt.style.context('ggplot') :
    hist_and_line()
```

```python
with plt.style.context('bmh'): #Bayesian Methods for Hackers style
    hist_and_line()
```

```python
with plt.style.context('dark_background'):
    hist_and_line()
```
